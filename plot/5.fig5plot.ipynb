```r
library(Seurat)
library(scop)
library(Signac)
library(ggplot2)
set.seed(4180)
setwd("/")
#########color
cols <- c("#444576", "#4682B4",
 "#AEDEEE","#FFA500",
 "#FFD790","#C65762","#FBDFDE",
 "#F6EFCF","#BCB99F")
pal <- colorRampPalette(cols)
```

```r
vCM <- readRDS("data/vCM.Rds")
DefaultAssay(vCM) <- "RNA"
```

```r
# fig5a
p <- FeatureDimPlot(vCM,features = 'G2M.Score',split.by = 'group',ncol = 2,theme_use = "theme_blank")
ggsave(
 plot = p, filename = "fig5a_vCM_G2M.pdf", path = "plot/figure5/",
 create.dir = TRUE, width = 5, height = 6
)
p
```

```r
# fig5b
p <- FeatureDimPlot(vCM,features = 'Ckap5',split.by = 'group',ncol = 2,theme_use = "theme_blank")
ggsave(
 plot = p, filename = "fig5b.pdf", path = "plot/figure5/",
 create.dir = TRUE, width = 5, height = 6
)
p
```

```r
# fig5c
p <- FeatureDimPlot(vCM,features = 'Ckap5',split.by = 'subtype',ncol = 2,theme_use = "theme_blank",cells.highlight = T)
ggsave(
 plot = p, filename = "fig5c.pdf", path = "plot/figure5/",
 create.dir = TRUE, width = 5, height = 5
)
p
```

```r
# fig5c
vCM$allgroup <- paste(vCM$subtype,vCM$group,sep = "_")
vCM$allgroup <- factor(paste(vCM$subtype,vCM$group,sep = '_'),levels = c('vCM-1_NP','vCM-1_EP','vCM-1_MP','vCM-1_LP','vCM-1_PP',
 'vCM-2_NP','vCM-2_EP','vCM-2_MP','vCM-2_LP','vCM-2_PP',
 'vCM-3_NP','vCM-3_EP','vCM-3_MP','vCM-3_LP','vCM-3_PP',
 'vCM-4_NP','vCM-4_EP','vCM-4_MP','vCM-4_LP','vCM-4_PP'))
p <- CellDensityPlot(
 vCM,
 features = "Ckap5",
 group.by = "allgroup",
 palcolor = rep(pal(4),each=5),
 y.min=0.01,legend.position = "none"
)
ggsave(
 plot = p, filename = "fig5d.pdf", path = "plot/figure5/",
 create.dir = TRUE, width = 3, height = 6
)
p
```
